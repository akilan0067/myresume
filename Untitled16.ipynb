```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Movie': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E'],
    'Release_Year': [2010, 2015, 2018, 2012, 2020],
    'IMDb_Rating': [7.5, 8.2, 6.7, 8.5, 7.0],
    'Box_Office_Collection (in Crores)': [100, 200, 150, 300, 120],
}

df = pd.DataFrame(data)
```

```python
(df)
```

```python
plt.bar(df['Movie'], df['IMDb_Rating'])
plt.xlabel('Movies')
plt.ylabel('IMDb Rating')
plt.title('IMDb Ratings of Indian Movies')
plt.show()
```

```python
plt.scatter(df['IMDb_Rating'], df['Box_Office_Collection (in Crores)'])
plt.xlabel('IMDb Rating')
plt.ylabel('Box Office Collection (in Crores)')
plt.title('IMDb Ratings vs Box Office Collection')
plt.show()
```
