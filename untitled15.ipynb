```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = pd.read_csv("/content/tested.csv")
```

```python
(titanic_data.head())
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(x="Survived", data=titanic_data, palette="Set1")
plt.title("Count of Survivors on the Titanic")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
titanic_data['Pclass'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=["#FF9999", "#66B2FF", "#99FF99"])
plt.title("Distribution of Passengers by Class")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Age Distribution of Passengers on the Titanic")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.countplot(x='survived', data=titanic)
plt.title('Count of Survived and Not Survived Passengers')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Count of Survived and Not Survived Passengers by Class')
plt.show()
```

```python
sns.histplot(titanic['age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.show
```

```python
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box Plot of Fare by Class')
plt.show()
```

```python
correlation_matrix = titanic.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
