```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
```

```python
# Scatter plot
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=iris_df, palette='viridis')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# Bar graph
avg_sepal_length = iris_df.groupby('target')['sepal length (cm)'].mean()
species_names = iris.target_names
```

```python
plt.bar(species_names, avg_sepal_length, color='skyblue')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.title('Average Sepal Length for Each Species')
plt.show()
```

```python
# Histogram
sns.histplot(iris_df, x='petal length (cm)', kde=True, hue='target', palette='Set1', element='step', stat='density')
plt.title('Distribution of Petal Lengths')
plt.show()
```

```python
# box plot
sns.boxplot(x='target', y='petal width (cm)', data=iris_df, palette='pastel')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.title('Distribution of Petal Widths for Each Species')
plt.show()
```
