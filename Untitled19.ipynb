```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as mpl
```

```python
df = pd.read_csv("/content/IRIS.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df["species"].unique()
```

```python
species_count = df['species'].value_counts()
```

```python
setosa = df[df["species"] == "Iris-setosa"]
setosa.head(5)
```

```python
versicolor = df[df["species"] == "Iris-versicolor"]
versicolor.head(5)
```

```python
sb.set(style = "dark")
```

```python
ax = sb.histplot(data = df, x = "sepal_length", hue = "species", multiple = "stack")
ax.set_title("Frequency of Different Sepal Length for Different Species")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Frequency")
mpl.show()
```

```python
ax = sb.boxplot(data = df, x = "species", y = "sepal_length")
ax.set_title("Box Plot for Sepal Length")
ax.set_xlabel("Species of Iris Flower")
ax.set_ylabel("Sepal Length")
mpl.show()
```

```python
virginica = df[df["species"] == "Iris-virginica"]
virginica.head(5)
```

```python
ax = sb.stripplot(data=df, x="species", y="sepal_length", hue = "species", jitter=True)
ax.set_title("Sepal Length Distribution by Species")
ax.set_xlabel("Species")
ax.set_ylabel("Sepal Length")
mpl.show()
```

```python
ax = sb.barplot(data=df, x="species", y="sepal_length", errorbar="sd")
ax.set_title("Mean Sepal Length by Species")
ax.set_xlabel("Species")
ax.set_ylabel("Mean Sepal Length")
mpl.show()
```
